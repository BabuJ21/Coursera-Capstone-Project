```python
import numpy as np
import pandas as pd
df=pd.read_csv('https://cocl.us/datascience_survey_data',index_col=0)

df1 = df.sort_values(['Very interested'],ascending=False)

```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
df2 = ((df1 / 2233) * 100).round(2)

df3=df2.plot(kind='bar', figsize=(20, 8),width=0.8,rot=90,color=['#5cb85c', '#5bc0de', '#d9534f'],fontsize=14) 

plt.title("Percentage of Respondents' interest in Data Science Areas",fontsize=16)
plt.legend(labels=df1.index, loc='upper right',fontsize=14)

plt.xticks(fontsize=14)
for spine in plt.gca().spines.values():
    spine.set_visible(False)
plt.yticks([])
for p in df3.patches:
    df3.annotate(str(p.get_height()) + '%', (p.get_x() * 1.005, p.get_height() * 1.03))
    
plt.show()
```
